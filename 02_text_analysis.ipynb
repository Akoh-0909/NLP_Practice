1. 적절한 데이터셋을 찾거나 생성하고, 적절한 전처리를 진행한다. [01_preprocessing.ipynb]
2. TF-IDF Vectorixer를 이용하여 벡터화한다.`TfidfVectorizer`
3. `cosine_similarity` 를 계산하여 입력된 문자열의 긍/부정을 판단한다.

##### 1. CSV 로드

```python
# CSV 로드
import pandas as pd

from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv('yelp_preprocessed.csv')


print(f'데이터 크기: {df.shape}')
df[['review', 'cleaned_review', 'label']].head()
```

```python
print(df['cleaned_review'])
print()
print(df['cleaned_review'].isna().sum())
```

##### 2. TF-IDF Vectorizer 로 벡터화

```python
# TF-IDF Vectorizer 로 벡터화
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
review_vecs = vectorizer.fit_transform(df['cleaned_review'])

print(review_vecs.shape)
print('학습된 단어 수:', len(vectorizer.get_feature_names_out()))
print("단어 사전:", vectorizer.vocabulary_)

display(review_vecs.toarray())
vectorizer.get_feature_names_out()

```

3. document similarity - Cosine Similarity 계산 + `cosine_similarity` 를 계산하여 입력된 문자열의 긍/부정을 판단

```python
# 입력 문자열과 전체 리뷰 간 Cosine Similarity 계산
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

similarities = cosine_similarity(review_vecs)
doc_arr = review_vecs.toarray()
cosine_similarity(doc_arr[1:2], doc_arr)

# index 0: 0.059  → 1번 리뷰와 약간 유사
# index 1: 1.000  → 자기 자신이라 유사도 1.0 (완전 동일)
# index 2: 0.000  → 3번 리뷰와 전혀 다름
# index 9: 0.201  → 10번 리뷰("Waited 30 minutes...")와 가장 유사
```

```python
# 유사한 리뷰의 label 기반으로 긍/부정 판단
import re
import nltk 
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

en_stopwords = stopwords.words('english')
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = text.lower()                                          # 소문자 변환
    text = re.sub(r'[^a-z0-9\s]', '', text)                     # 특수문자 제거
    tokens = word_tokenize(text)                                 # 토큰화
    en_stopwords = stopwords.words('english')
    tokens = [t for t in tokens if t not in en_stopwords]       # 불용어 제거
    tokens = [t for t in tokens if len(t) > 2]                  # 짧은 단어 제거
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t, pos='v') for t in tokens] # 표제어 추출
    return ' '.join(tokens)

```

```python
def predict_sentiment(input_text):
    cleaned_input = preprocess(input_text)
    print(f'전처리 결과: "{cleaned_input}"')

    input_vec = vectorizer.transform([cleaned_input])
    similarities = cosine_similarity(input_vec, review_vecs).flatten()

    most_similar_idx = similarities.argmax()
    label = df['label'][most_similar_idx]
    print(f'가장 유사한 리뷰: {df["review"][most_similar_idx]}')
    if label == 1:                      # 1 = 긍정
        print('최종 판단: 긍정')
    else:                               # 0 = 부정
        print('최종 판단: 부정')
```

```python
predict_sentiment("The food was delicious and staff was very friendly!")
predict_sentiment("Terrible food and the service was really slow.")
```

```python
# 테스트 1: 긍정 리뷰
print('=' * 60)
print('입력:', "The food was delicious and the staff was very friendly!")
print('=' * 60)
predict_sentiment("The food was delicious and the staff was very friendly!")
```

```python
# 테스트 2: 부정 리뷰
print('=' * 60)
print('입력:', "Terrible food and the service was really slow and disappointing.")
print('=' * 60)
predict_sentiment("Terrible food and the service was really slow and disappointing.")
```

```python
# 테스트 3: 직접 입력해보기
my_review = "The pasta was great but the service was a bit slow."
print('=' * 60)
print('입력:', my_review)
print('=' * 60)
predict_sentiment(my_review)
```
